```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid', palette='muted', font_scale=1.5)
pd.set_option("max_rows", 100)
pd.set_option("max_columns", None)
```

```python
#Regression Metrics
import math
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
def regression_metrics(regressor, X_train, X_test, y_train, y_test):
    # Metrics Evaluation
    y_test_pred = np.expm1(regressor.predict(X_test))
    y_train_pred = np.expm1(regressor.predict(X_train))
    # Metrics on training 
    mse = mean_squared_error(y_train, y_train_pred)
    rmse = math.sqrt(mse)
    mae = mean_absolute_error(y_train, y_train_pred)
    r2s = r2_score(y_train, y_train_pred)
    print("***** Training Metrics *****")
    print("Mean Squared Error: ",mse)
    print("Root Mean Squared Error: ",rmse)
    print("Mean Absolute Error: ",mae)
    print("R2 Score: ",r2s)
    # Metrics on testing
    mse = mean_squared_error(y_test, y_test_pred)
    rmse = math.sqrt(mse)
    mae = mean_absolute_error(y_test, y_test_pred)
    r2s = r2_score(y_test, y_test_pred)
    print("\n***** Testing Metrics *****")
    print("Mean Squared Error: ",mse)
    print("Root Mean Squared Error: ",rmse)
    print("Mean Absolute Error: ",mae)
    print("R2 Score: ",r2s)
```

```python
df = pd.read_csv('pre-processed-train-data.csv')
df
```

```python
X = df.drop(['SalePrice', 'Id'], axis=1)
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print("X_train: {}".format(X_train.shape))
print("X_test: {}".format(X_test.shape))
print("y_train: {}".format(y_train.shape))
print("y_test: {}".format(y_test.shape))
```

```python
# XGBoost
import xgboost
xgb_model = xgboost.XGBRegressor(
    colsample_bynode=0.1,
    gamma = 0,
    learning_rate=0.018,
    max_depth=8,
    n_estimators = 2500,
    reg_alpha=0.75,
    reg_lambda=0.25,
    subsample=0.5,
    seed=0,
)
xgb_model.fit(X, np.log1p(y))
```

```python
#Regular to submit
df_test = pd.read_csv('pre-processed-test-data.csv')
input_test = df_test.drop(['Id'], axis=1)

 # pred = regressor.predict(input_test)
 # pred = xgb_model.predict(input_test)
pred = np.expm1(xgb_model.predict(input_test))

submission = pd.concat([df_test[['Id']] ,pd.DataFrame(pred, columns=['SalePrice'])], axis=1)
submission.to_csv('submission.csv', index=False)
```
